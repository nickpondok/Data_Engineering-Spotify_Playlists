The goal of this project is to do an analysis on spotify curated playlists. The hope would be to see if there are any trends amongs playlists from the same genre.

I was able to build a pipeline that worked for a sample genre of music. In the pipline functions were created to do the following:

1. Given a genre_id (i.e. pop,hiphop,country) all spotify curated playlists are returned 
2. Given a playlist ID all tracks on the playlist are returned
3. given a track all audio features are returned 

Final data frames are passed into my sql database. Any dataframes that will be plotted are saved to csvs so they can be pulled into my steamlit python file. I was able to push through the values for the hiphop genre to the steamlit app. 

Next steps will be to test the pipeline for more and more data to ensure it handles it all well. I also hope to add more interactivity to the the streamlit app in order to take in input so the user can pick which genres they are interested in and potentially look at specific playlists.
